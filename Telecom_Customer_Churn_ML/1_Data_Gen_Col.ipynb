```python
%pip install faker
%pip install tqdm
import pandas as pd
import numpy as np
from faker import Faker
from tqdm import tqdm

# Load your existing dataset
file_path = "/Volumes/kusha_solutions/telecom_churn_ml/my_volume/customers_data/Telco-Customer-Churn.csv"
df = pd.read_csv(file_path)

# Initialize Faker
fake = Faker()

# Number of new rows
num_records = 500000

# Containers
new_data = []

for _ in tqdm(range(num_records), desc="Generating telecom data"):
    gender = np.random.choice(["Male", "Female"])
    senior = np.random.choice([0, 1], p=[0.85, 0.15])
    partner = np.random.choice(["Yes", "No"])
    dependents = np.random.choice(["Yes", "No"])
    tenure = np.random.randint(0, 73)
    phone_service = np.random.choice(["Yes", "No"])
    multiple_lines = "Yes" if phone_service == "Yes" and np.random.rand() > 0.6 else "No"
    internet_service = np.random.choice(["DSL", "Fiber optic", "No"])
    online_security = np.random.choice(["Yes", "No", "No internet service"]) if internet_service != "No" else "No internet service"
    online_backup = np.random.choice(["Yes", "No", "No internet service"]) if internet_service != "No" else "No internet service"
    device_protection = np.random.choice(["Yes", "No", "No internet service"]) if internet_service != "No" else "No internet service"
    tech_support = np.random.choice(["Yes", "No", "No internet service"]) if internet_service != "No" else "No internet service"
    streaming_tv = np.random.choice(["Yes", "No", "No internet service"]) if internet_service != "No" else "No internet service"
    streaming_movies = np.random.choice(["Yes", "No", "No internet service"]) if internet_service != "No" else "No internet service"
    contract = np.random.choice(["Month-to-month", "One year", "Two year"])
    paperless_billing = "Yes" if contract == "Month-to-month" else np.random.choice(["Yes", "No"])
    payment_method = np.random.choice([
        "Electronic check", "Mailed check", "Bank transfer (automatic)", "Credit card (automatic)"
    ])
    monthly_charges = np.round(np.random.uniform(18, 120), 2)
    total_charges = np.round(monthly_charges * tenure + np.random.uniform(0, 100), 2)
    customer_id = fake.uuid4()[:8]

    # üéØ Churn logic
    # Customers more likely to churn:
    # - Month-to-month contracts
    # - High monthly charges
    # - Low tenure
    # Medium churn (‚ÄúMaybe‚Äù): mid-level tenure and charges
    # Stable (‚ÄúNo‚Äù): long tenure or multi-year contracts
    if contract == "Month-to-month" and tenure < 12 and monthly_charges > 80:
        churn = "Yes"
    elif 12 <= tenure <= 30 and 50 <= monthly_charges <= 90:
        churn = "Maybe"
    else:
        churn = "No"

    new_data.append([
        customer_id, gender, senior, partner, dependents, tenure, phone_service, multiple_lines,
        internet_service, online_security, online_backup, device_protection, tech_support,
        streaming_tv, streaming_movies, contract, paperless_billing, payment_method,
        monthly_charges, total_charges, churn
    ])

# Build dataframe
new_df = pd.DataFrame(new_data, columns=df.columns)

# Append to original data
combined_df = pd.concat([df, new_df], ignore_index=True)

# Save back to same file
combined_df.to_csv(file_path, index=False)

print("‚úÖ 5 Lakh new records added successfully!")
print("Class distribution after update:\n", combined_df["Churn"].value_counts())
```
